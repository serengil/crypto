```python
import math
```

# key generation

```python
def is_prime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```

```python
p = 11; q = 13
```

```python
assert is_prime(p)
assert is_prime(q)
```

```python
n = p * q
phi = (p-1)*(q-1)
print(f"totient({n}) = {phi}")
```

```python
# private key - pick a random integer
e = 7
assert math.gcd(e, phi) == 1
```

```python
d = pow(e, -1, phi)
assert (e * d) % phi == 1
```

# multiplicative homomorphic features

```python
m1 = 9
m2 = 11
```

```python
m1_encrypted = pow(m1, e, n)
m2_encrypted = pow(m2, e, n)
```

```python
assert (m1_encrypted * m2_encrypted) % n == pow(m1*m2, e, n) % n
```
